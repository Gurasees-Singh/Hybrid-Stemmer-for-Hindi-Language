```python
import nltk
import collections
from nltk.corpus import PlaintextCorpusReader
from nltk.corpus import indian
hindi_word = nltk.corpus.indian.words('hindi.pos')
hindi_word = list(hindi_word)
a=[]
```

```python
for i in range(len(hindi_word)):
    temp=hindi_word[i]
    g=len(temp)
    if(g>=1):
        temp1=temp[-1:]
    else:
        temp1='null'
    if(g>=2):
        temp2=temp[-2:]
    else:
        temp2='null'
    if(g>=3):
        temp3=temp[-3:]
    else:
        temp3='null'
    if(g>=4):
        temp4=temp[-4:]
    else:
        temp4='null'
    if(g>=5):
        temp5=temp[-5:]
    else:
        temp5='null'
    a.append([temp,temp1,temp2,temp3,temp4,temp5])
```

```python
list1=[]
list2=[]
list3=[]
list4=[]
list5=[]

for i in range(len(a)):
    if(a[i][1]!='null'):
        list1.append(a[i][1])
    if(a[i][2]!='null'):
        list2.append(a[i][2])
    if(a[i][3]!='null'):
        list3.append(a[i][3])
    if(a[i][4]!='null'):
        list4.append(a[i][4])
    if(a[i][5]!='null'):
        list5.append(a[i][5])
```

```python
f1=collections.Counter(list1)
f2=collections.Counter(list2)
f3=collections.Counter(list3)
f4=collections.Counter(list4)
f5=collections.Counter(list5)

suffix=[]
for i,j in f1.items():
    if(j>100):
        suffix.append(i)

for i,j in f2.items():
    if(j>100):
        suffix.append(i)

for i,j in f3.items():
    if(j>100):
        suffix.append(i)

for i,j in f4.items():
    if(j>100):
        suffix.append(i)

for i,j in f5.items():
    if(j>100):
        suffix.append(i)
```
